# Calling the make regression utility of scikit learn to create a synthetic data set of 1 feature
# Then convert into a tensor

```python
from sklearn.datasets import make_regression
```

```python
x,y=make_regression(30,1)
```

```python
y= y.reshape(30,1)
```

```python
x=torch.tensor(x, dtype=torch.float32)
y=torch.tensor(y, dtype= torch.float32)
```

```python
x
```

```python
from  torch import nn
```

```python
class model(nn.Module):
    def __init__(self):
        super(model, self).__init__()
        self.layer = nn.Linear(1, 1)
        
    def forward(self, x):
        x = self.layer(x)
        return x
```

```python
m=model()
```

```python
m.parameters
```

```python
define_criterion = torch.nn.MSELoss()

SGD_optimizer = torch.optim.SGD(m.parameters(), lr=0.1)
```

```python
for epoch in range(50): 

  
    predict_y = m(x) 
    loss = define_criterion(predict_y, y) 

  
     
    loss.backward() 

    SGD_optimizer.step() 
    SGD_optimizer.zero_grad()
    print('epoch {}, loss function {}'.format(epoch, loss.item()))
```

```python
plt.scatter(x,y)
plt.plot(x, m(x).detach())
```

# The muti Label Example

```python
x2,y2= make_regression(100, 1, n_targets=2)
```

```python
y2
```

```python
x2=torch.tensor(x2, dtype=torch.float32)
y2=torch.tensor(y2, dtype= torch.float32)
```

```python
x2
```

```python
y2
```

```python
class model2(nn.Module):
    def __init__(self):
        super(model2, self).__init__()
        self.layer = nn.Linear(1, 2)
        
    def forward(self, x):
        x = self.layer(x)
        return x
```

```python
m2=model2()
```

```python
define_criterion = torch.nn.MSELoss()

SGD_optimizer = torch.optim.SGD(m2.parameters(), lr=0.1)
```

```python
for epoch in range(500): 

  
    predict_y2 = m2(x2) 
    loss2 = define_criterion(predict_y2, y2) 

  
     
    loss2.backward() 

    SGD_optimizer.step() 
    SGD_optimizer.zero_grad()
    print('epoch {}, loss function {}'.format(epoch, loss2.item()))
```

```python
m2(x2)
```
